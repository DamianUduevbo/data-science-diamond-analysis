## PHASE 7:
Using a 2-sample t-test to determine if there is a statistical difference between the price of diamonds with a cut that is considered:
* 'Good' vs 'Very Good'
* 'Premium' vs 'Ideal'
* set alpha to 5% i.e. 0.05.

```python
price_good = df[df['cut'] == 'Good']['price']
price_very_good = df[df['cut'] == 'Very Good']['price']
price_premium = df[df['cut'] == 'Premium']['price']
price_ideal = df[df['cut'] == 'Ideal']['price']

alpha = 0.05

# 'Good' vs 'Very Good'
t_stat_goodness, p_value_goodness = stats.ttest_ind(price_good, price_very_good)

if p_value_goodness < alpha:
  print("The price of 'Good' and 'Very Good' cuts is statistically different.")
else:
  print("The price of 'Good' and 'Very Good' cuts is not statistically different.")

# 'Premium' vs 'Ideal'
t_stat_prem, p_value_premium = stats.ttest_ind(price_premium, price_ideal)
if p_value_premium < alpha:
  print("The price of 'Premium' and 'Ideal' cuts is statistically different.")
else:
  print("The price of 'Premium' and 'Ideal' cuts is not statistically different.")
```
